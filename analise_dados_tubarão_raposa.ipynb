```python
#apagando colunas pratimente nulas
df = df.drop(columns=[
    "mediaType",
    "establishmentMeans",
    "typeStatus",
    "dateIdentified",
    "recordNumber",
    "depthAccuracy",
    "elevationAccuracy",
    "elevation",
    "coordinatePrecision",
    "infraspecificEpithet"
])
```

```python
df.nunique()
```

```python
paises = {
    'CL': 'Chile', 'ID': 'Indonésia', 'US': 'Estados Unidos',
    'OM': 'Omã', 'NG': 'Nigéria', 'IN': 'Índia',
    'FO': 'Ilhas Faroe', 'CI': 'Costa do Marfim', 'NO': 'Noruega',
    'GH': 'Gana', 'GB': 'Reino Unido', 'MX': 'México',
    'AU': 'Austrália', 'GA': 'Gabão', 'SL': 'Serra Leoa'
}

tabela = (df["countryCode"]
          .value_counts()
          .head(15)
          .reset_index()
          .rename(columns={"countryCode": "País", "count": "Registros"}))

tabela["País"] = tabela["País"].map(paises)

tabela #primeiros 15 paises registrados (mudar dps para um top algum valor )

```
